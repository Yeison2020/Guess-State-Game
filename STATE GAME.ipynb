```python
from turtle import Turtle, Screen
import turtle
import pandas
screen = Screen()

screen.title("U.S States Game")
#screen.setup(725, 421)

image = "blank_states_img.gif"

screen.addshape(image)

game = Turtle()
game.shape(image)
# Another alternative to keep my screen open all the neccesary time needed


# Code to get the locations coor() of the states
"""
def get_mouse_click_coor(x,y):
    
    print(x,y)
    
    
turtle.onscreenclick(get_mouse_click_coor)

"""



# Step 1 pandas reads 

data = pandas.read_csv("50_states.csv")
all_states = data.state.to_list()
guessed_States =[]



# Check if the answer is one of the states in data: if they got it right create a turtle to write the name of the state in the map screen


while len(guessed_States) < 50:
    

    answer_state = screen.textinput(title=f"{len(guessed_States)}/50 States Corrects", prompt="What is the state ?").title()
    if answer_state == "Exit":
        missing_states = []
        for state in all_states:
            if state not in guessed_States:
                missing_states.append(state)
        new_data = pandas.DataFrame(missing_states)
        new_data.to_csv("States_to_learn.csv")
        break
    if answer_state  in all_states:
        guessed_States.append(answer_state)
        t = Turtle()
        t.hideturtle()
        t.penup()
        state_data = data[data.state == answer_state]
        # Beutiful extracting the data from the csv file with the pandas x and y, because the pandas modules allows me to got directly 
        t.goto(int(state_data.x), int(state_data.y))
        t.write(answer_state)




    
    
    
screen.mainloop()
```
